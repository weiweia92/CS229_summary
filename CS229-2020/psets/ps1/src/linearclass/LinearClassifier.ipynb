$$\mu_{y^{(i)}}=\text{1}\{y^{(i)}=0\}\mu_0+\text{1}\{y^{(i)}=1\}\mu_1$$

$$p(x^{(i)}|y^{(i)};\mu_0,\mu_1,\Sigma)=\frac{1}{(2\pi)^{n/2}|\Sigma|^{1/2}}\text{exp}\big\{-\frac{1}{2}(x^{(i)}-\mu_{y^{(i)}})^T\Sigma^{-1}(x^{(i)}-\mu_{y^{(i)}})\big\}$$

$$p(y^{(i)};\phi)=\phi^{\text{1}\{y^{(i)}=1\}}(1-\phi)^{\text{1}-\text{1}\{y^{(i)}=1\}}$$

To compute $\phi,\mu_0$ and $\mu_1$, recall the log-likelihood:

$$\begin{equation*} 
    \begin{split}
        &\ell(\phi,\mu_0,\mu_1,\Sigma) \\
        &=\sum_{i=1}^m\text{log}p(x^{(i)}|y^{(i)};\phi,\mu_0,\mu_1,\Sigma)+\sum_{i=1}^m\text{log}p(y^{(i)};\phi)\\
        &=\sum_{i=1}^m\text{log}\frac{1}{(2\pi)^{n/2}|\Sigma|^{1/2}}\text{exp}\big\{-\frac{1}{2}(x^{(i)}-\mu_{y^{(i)}})^T\Sigma^{-1}(x^{(i)}-\mu_{y^{(i)}})\big\}+\sum_{i=1}^m\text{log}\phi^{\text{1}\{y^{(i)}=1\}}(1-\phi)^{\text{1}-\text{1}\{y^{(i)}=1\}} \\
        &=-\frac{mn}{2}\text{log}(2\pi)-\frac{m}{2}\text{log}|\Sigma|-\frac{1}{2}\sum_{i=1}^m(x^{(i)}-\mu_{y^{(i)}})^T\Sigma^{-1}(x^{(i)}-\mu_{y^{(i)}})+\sum_{i=1}^m \text{1}\{y^{(i)}=1\}\frac{1}{\phi}+\sum_{i=1}^m \text{1}\{y^{(i)}=0\}\frac{1}{1-\phi}
    \end{split}
\end{equation*}$$

Take derivative of $\ell$ w.r.t $\phi$ and set to 0: 

$$\frac{\partial}{\partial \phi}\ell(\phi,\mu_0,\mu_1,\Sigma)=\sum_{i=1}^m \text{1}\{y^{(i)}=1\}\frac{1}{\phi}+\sum_{i=1}^m \text{1}\{y^{(i)}=0\}\frac{1}{1-\phi}\tag{1}$$

$$\because \sum_{i=1}^m \text{1}\{y^{(i)}=0\}=m-\sum_{i=1}^m \text{1}\{y^{(i)}=1\}$$

$$\therefore (1)\text{式}=\sum_{i=1}^m \text{1}\{y^{(i)}=1\}\frac{1}{\phi}+(m-\sum_{i=1}^m \text{1}\{y^{(i)}=1\})\frac{1}{1-\phi}=0$$

So we have $$\phi=\frac{1}{m}\sum_{i=1}^m \text{1}\{y^{(i)}=1\}$$

Also, take derivative of $\ell$ w.r.t $\mu_0$ and set to 0: 

$$\frac{\partial}{\partial \mu_0}\ell=\sum_{i=1}^m \text{1}\{y^{(i)}=0\}\Sigma^{-1}(x^{(i)}-\mu_0)=0$$

We can easily obtain that $\mu_0=\sum_{i=1}^m\text{1}\{y^{(i)=0}\}x^{(i)}/\sum_{i=1}^m\text{1}\{y^{(i)}=0\}$. Similarly $\mu_1=\sum_{i=1}^m\text{1}\{y^{(i)=1}\}x^{(i)}/\sum_{i=1}^m\text{1}\{y^{(i)}=1\}$

$$\frac{\partial}{\partial \Sigma}\ell(\phi,\mu_0,\mu_1,\Sigma)=-\frac{m}{2\Sigma}+\frac{1}{2}\sum_{i=1}^m(x^{(i)}-\mu_{y^{(i)}})(x^{(i)}-\mu_{y^{(i)}})^T\Sigma^{-2}=0$$

We have $$\Sigma=\frac{1}{m}\sum_{i=1}^m(x^{(i)}-\mu_{y^{(i)}})(x^{(i)}-\mu_{y^{(i)}})^T$$
