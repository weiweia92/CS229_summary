**1.(a)**

Our model converges on dataset A but fails to converge on dataset B.

**(b)**



Let $\theta$ be a parameter vector such that dataset B is completely separated by the hyperplane $\theta^Tx=0$. Now, consider a parameter $\theta'=c\cdot \theta$. Given an example $x$, we observe that:

* If $y=0$, then $\theta^T x <0$ and its loss $\text{w.r.t}$ $\theta'$ is: $-\text{log}\big( \frac{1}{1+\text{exp}(-c \cdot \theta^T x)} \big)$

* If $y=1$, then $\theta^T x >0$ and its loss $\text{w.r.t}$ $\theta'$ is: $-\text{log}\big(1- \frac{1}{1+\text{exp}(-c \cdot \theta^T x)} \big)$

As $c\rightarrow \infty$, in both cases the losses are strictly decreasing without a bound. And so the total cost is a lower-unbounded convex function. The Gradient Descent will fail to converge to the global minimum since there is not such one. Due to inseparability, dataset A doesn't endure the above trait, our optimizer actually worked nicely.

**(c)**

**i. Using a different constant learning rate.**

No. Modifying the selection scheme for learning rates doesn't change the linear separability, hence the model wouldn't converge.

**ii. Decreasing the learning rate over time (e.g. scaling the initial learning rate by $1/t^2$, where $t$ is the number of gradient descent iterations thus far).**

No. Same answer as (i)

**iii. Linear scaling of the input features.**

No. Linear transformation doesn't remove the separability of dataset B, hence the model still wouldn't converge.

**iv. Adding a regularization term $\parallel \theta \parallel_2^2$ to the loss function.**

Yes. Regularization can help in this case. The cost function is modified and might have a minimum now.

**v. Adding zero-mean Gaussian noise to the training data or labels.**

Adding noise terms might break the separability and make the model converge.

**(d)**

The Support Vector Machine is not vulnerable to separable datasets like B. The hard margin SVM actually dedicates to solve this problem. It tries to find a maximum margin classifier for such a dataset.
