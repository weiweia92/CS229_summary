**(b) Compression Factor**

Originally, each pixel in our image is represented by 24 bits. If we compress the image with 16 colors, each pixels now only uses 4 bits ($2^4=16$). Therefore, the image is compressed roughly by a factor of $4/24 \approx 1/6$
